```python
import random
import numpy as np
import matplotlib.pyplot as plt
p = 0.5;p_tilde = 0.5
count = 0; nsamples = 1000; tmax = 10000
for n in range(1, nsamples+1):
    print("Sample: ", n)
    positions = [0]
    r1 = np.random.random(tmax)
    r2 = np.random.random(tmax)
    upp = r1 < p
    upp_tilde = r2 < p_tilde
    for k in range(1, tmax):
        repeatedNum = np.sum(positions == positions[-1])
        if ((max(positions) == positions[-1]) & (repeatedNum == 1)):
            up = (upp_tilde[k-1] == 1)
            down = ((upp_tilde[k-1] == 0) & (positions[-1] != 0))
        else:
            up = (upp[k-1] == 1)
            down = ((upp[k-1] == 0) & (positions[-1] != 0))
        Sk = positions[-1] - down + up
        if Sk == 0:
            count += 1
            print("Return to zero within tmax.")
            break
        positions.append(Sk)
print("Probability of return to zero within tmax:",count / nsamples)
```
